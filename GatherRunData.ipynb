```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import openpyxl
import os
import pandas as pd
import seaborn as sns
```

```python
# gets the monthly report
os.chdir('/Users/zacharyellis/Downloads/')
monthlyReport = 'Sept9_run_report.xlsx'
df = pd.read_excel(monthlyReport)
```

```python
df['start_date'] = pd.to_datetime(df['start_date'])
df = df.sort_values(by='start_date',ascending=False)
df
```

# 3) Plotting Data

```python
df.to_numpy
elevation_number = df.loc[: , 'elevation_gain'].values
distance_number = df.loc[:, 'distance'].values
hours_number = df.loc[: , 'hours'].values


plt.plot(distance_number, elevation_number)

plt.savefig('books_read3.png')
```

```python
plt.scatter(distance_number, elevation_number, c='g', label = 'y = 17.5622x - 0.3662')



plt.ylim(bottom=0, top=650)  # this line
plt.xlim(left=0, right=35)  # this line

plt.ylabel('Elevation Gain (ft)', fontsize=15)
plt.xlabel('Distance in Mi', fontsize=15)
plt.title('Strava Runs', fontsize=20)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


plt.style.use('seaborn')
plt.legend()
plt.show()

plt.tight_layout()
plt.savefig('books_read4.png')
```
