# Pie Chart

```python
import matplotlib.pyplot as plt
num_activities = [29, 16, 27, 42, 35, 67, 
                  52, 46, 76, 64, 38, 13]

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
         'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# explode the most active month 
exp = [0, 0, 0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0]
cl = ['Lime', 'Violet', 'Teal', 'Tan']
plt.pie(num_activities, labels = months, explode = exp, colors = cl, autopct= '%2.1f%%')
plt.show()
```
