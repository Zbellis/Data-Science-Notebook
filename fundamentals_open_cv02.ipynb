```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Read/Display an image

```python
image_color = mpimg.imread("6.png")
```

```python
plt.imshow(image_color)
```

```python
image_color.shape
```

# Convert an image into Grayscale

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
```

```python
image_gray.shape
```

```python
image_gray
```

# Save the grayscale image

```python
# Save the image in greyscale
cv2.imwrite("6_grayscale.png", image_gray)
```

```python
color_image = cv2.imread("6.png")

 

cartoon_image = cv2.stylization(color_image, sigma_s = 200, sigma_r = 0.3)
# cartoon_image, cartoon_image2 = cv2.pencilSketch(color_image, sigma_s=60, sigma_r=0.1)
```
